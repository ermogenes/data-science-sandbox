```python
import sklearn as sl
from sklearn import tree

type(sl)
```

* **Feature**: Atributos das lições
* **Label**: Classificação da lição

Ex.: São _features_ de uma possível operação de crédito o [valor solicitado] e o [índice de idoneidade do cliente]. Seus labels possíveis são [Emprestar] e [Não emprestar].

```python
features = [[1000.00, 0.5], [50000.00, 0.9], [15000.00, 0.3], [1000.00, 0.9], [5000.00, 0.65], [200000.00, .95]]
labels = [1, 1, 0, 1, 0, 0]
```

## Treinando um classificador

```python
classificador = tree.DecisionTreeClassifier()
classificador
```

```python
classificador = classificador.fit(features, labels)
```

## Fazendo uma classificação

```python
classificador.predict([[90000.00, 0.15]])
```

Uma predição bem ruim, de fato. Precisamos de mais dados.

## Usando um banco de dados acadêmico

```python
from sklearn.datasets import load_iris
```

http://scikit-learn.org/stable/datasets/index.html#datasets

```python
iris = load_iris()

type(iris)
```

### Navegando pelo dataset

### Separando o dataset em dados de treinamento e dados de verificação

```python
import numpy as np

linhas_teste = [0, 26, 50, 69, 100, 147]

features_teste = iris.data[linhas_teste]
labels_teste = iris.target[linhas_teste]

features_treinamento = np.delete(iris.data, linhas_teste, axis=0)
labels_treinamento = np.delete(iris.target, linhas_teste)
```

```python
clf_iris = tree.DecisionTreeClassifier()
clf_iris = clf_iris.fit(features_treinamento, labels_treinamento)
```

### Comparando o resultado

```python
labels_teste
```

```python
predicoes = clf_iris.predict(features_teste)

predicoes
```

### Visualizando as regras inferidas

* A visualização em grafo usa o pydotplus (https://anaconda.org/anaconda/pydotplus);
* É necessária a instalação física do GraphViz (http://www.graphviz.org/) e sua inclusão no PATH;
* Pode ser necessária a instalação do graphviz (https://anaconda.org/anaconda/graphviz);
* Nesse link há um registry file que pode ser necessário também: (https://github.com/erocarrera/pydot/issues/65).

```python
from IPython.display import Image 
from sklearn.externals.six import StringIO
import pydotplus

dot_data = StringIO()

tree.export_graphviz(
    clf_iris,
    out_file = dot_data,
    feature_names = iris.feature_names,
    class_names = iris.target_names,
    filled = True,
    rounded = True,
    impurity = False
)

grafo = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(grafo.create_png()) 
```

## Medindo a acurácia do modelo

```python
from sklearn.metrics import accuracy_score

accuracy_score(labels_teste, predicoes)
```
